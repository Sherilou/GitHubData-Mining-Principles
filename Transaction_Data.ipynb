```
import pandas as pd

dataset1 = pd.read_csv('Transaction_Data (1).csv')
```

```
print(dataset1.head())
```

```
print(dataset1.isnull().sum())
```

```
dataset1 = dataset1.drop(columns=['Transaction_Amount'])
```

```
dataset1 = dataset1.drop_duplicates()
```

```
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

dataset1['Transaction_Type'] = label_encoder.fit_transform(dataset1['Transaction_Type'])

dataset1
```

Label Encoding has been applied to the 'Transaction_Type' column.

```
dataset1['Transaction_Hour'] = pd.to_datetime(dataset1['Transaction_Date']).dt.hour

dataset1['Transaction_DayOfWeek'] = pd.to_datetime(dataset1['Transaction_Date']).dt.dayofweek


dataset1['Transaction_WeekPart'] = dataset1['Transaction_DayOfWeek'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

dataset1['Transaction_WeekPart'] = label_encoder.fit_transform(dataset1['Transaction_WeekPart'])

dataset1
```

New features have been created: 
1. `Transaction_Hour`: Extracted hour of the transaction.
2. `Transaction_DayOfWeek`: Day of the week (0=Monday, 6=Sunday).
3. `Transaction_WeekPart`: Categorized as 'Weekday' or 'Weekend' (encoded).

```
# Splitting the dataset into features (X) and target (y)
X = dataset1.drop(columns=['Transaction_Type'])
y = dataset1['Transaction_Type']

X, y
```

The dataset has been successfully split into features (X) and target (y).
